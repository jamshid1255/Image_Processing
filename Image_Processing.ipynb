### Grayscale and Color Images

```python
# generate image and save in storage
import os


randomArray = bytearray(os.urandom(120000))
print(randomArray)
flatNumpyArray = np.array(randomArray)
print(flatNumpyArray)

# convert flat array to AxB grayscale Image.

grayImage = flatNumpyArray.reshape(300,400)
cv2_imshow(grayImage)
cv2.imwrite('/content/drive/MyDrive/Image Processing/mygrayImageImage.png',grayImage )

print('\n')
# convert flat array to AxBxC Color Image.
BGRColorImage = flatNumpyArray.reshape(100,400,3)
cv2_imshow(BGRColorImage)
cv2.imwrite('/content/drive/MyDrive/Image Processing/myBGRColorImageImage.png',BGRColorImage )
```

### Resize the image

```python
lena = cv2.imread('/content/drive/MyDrive/Image Processing/images.jpg')
print(lena.shape)
image = lena
tenth = cv2.resize(lena, (10,20), fx = 0.1, fy = 0.1)
ten = cv2.resize(lena,(0,0), fx=0.1, fy=0.1)
v = cv2.resize(lena,(300,500),fx=0.1, fy=0.1)
bigger = cv2.resize(lena,(3000,2000))
stretch_near = cv2.resize(lena,(780,540),interpolation = cv2.INTER_NEAREST)
Title = ["Orignal","Tenth", "Bigger", "Interpolation Nearest"]
images = [lena,tenth, bigger,stretch_near]
cv2_imshow(tenth)
cv2_imshow(ten)
cv2_imshow(v)
cv2count = 4
Titles = ["Orignal","Half", "Bigger", "Interpolation Nearest"]
count = 4

for i in range(count):
  plt.subplot(2,2,i+1)
  plt.title(Titles[i])
  plt.imshow(cv2.cvtColor(images[i],cv2.COLOR_BGR2RGB))

plt.show()
```

### Reduce Intensity Level 

```python
img1 = cv2.imread('/content/drive/MyDrive/Image Processing/images.jpg')
print(img1.shape)
img2 = np.uint8(img1/128)
print(img2)
img3 = img2*255
print(img3)
cv2_imshow(img3)
norm_image1 = cv2.normalize(img2,None, 100 ,200, norm_type = cv2.NORM_MINMAX)
print(norm_image1)
cv2_imshow(norm_image1)
```
